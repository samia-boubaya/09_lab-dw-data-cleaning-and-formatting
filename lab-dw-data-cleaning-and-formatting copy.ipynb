```python
# Standardize columns names --------------------------------------------------------------
# Customer to customer_id
# ST to state
# GENDER to gender
# Education to education_level
# Customer Lifetime Value to lifetime_value
# Monthly Premium Auto to monthly_premium
# Number of Open Complaints to open_complaints
# Policy Type to policy_type
# Vehicle Class to vehicle_class
# Total Claim Amount to total_claim_amount
# ----------------------------------------------------------------------------------------

df.rename(columns={
    "Customer": "customer_id",
    "ST": "state",
    "GENDER": "gender",
    "Education": "education_level",
    "Customer Lifetime Value": "lifetime_value",
    "Income": "annual_income",
    "Monthly Premium Auto": "monthly_premium",
    "Number of Open Complaints": "open_complaints",
    "Policy Type": "policy_type",
    "Vehicle Class": "vehicle_class",
    "Total Claim Amount": "total_claim_amount"
}, inplace=True)
df.head(10)
```

## Exercise 2: Cleaning invalid Values

- The **dataset contains columns with** `inconsistent` and `incorrect values` that could **affect the accuracy of our analysis**. 
- Therefore, we need to **clean these columns to ensure that they only contain valid data**.

**Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).**

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", 
    - which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, 
    - for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
df.info()
```

---
### **gender**
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", 
    - which need to be standardized, for example, to "M" and "F".*

```python
df["gender"].unique()
```

```python
# mapping gender values
gender_mapping = {
    "female": "F",
    "Femal": "F",    
    "Male": "M"
}
# replace values in gender
df['gender'] = df['gender'].replace(gender_mapping)

# Check unique values after mapping
print(df['gender'].unique())
```

---
### **state**
- *State abbreviations be can replaced with its full name, 
    - for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*

```python
df["state"].unique()
```

```python
# mapping gender values
state_mapping = {
    "WA": "Washington",
    "AZ": "Arizona",    
    "Cali": "California"
}
# replace values in state
df['state'] = df['state'].replace(state_mapping)

# Check unique values after mapping
print(df['state'].unique())
```

---
### **education_level**
- *In education, "Bachelors" could be replaced by "Bachelor"*

```python
df["education_level"].unique()
```

```python
# mapping gender values
education_mapping = {
    "Bachelors": "Bachelor"
}
# replace values in education_level
df['education_level'] = df['education_level'].replace(education_mapping)

# Check unique values after mapping
print(df['education_level'].unique(), "\n")
df["lifetime_value"].info()
```

---
### **lifetime_value**
- *In Customer Lifetime Value, delete the `%` character*

```python
df["lifetime_value"].unique()
```

```python
df["lifetime_value"].info()
```

```python
# Remove '%' and convert to float
df['lifetime_value'] = df['lifetime_value'].str.replace('%', '', regex=False).astype(float)
```

```python
# Check the first few values
df['lifetime_value'].head(10)
```

---
### **vehicle_class**
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
df["vehicle_class"].unique()
```

```python
df["vehicle_class"].info()
```

```python
# mapping gender values
vehicle_mapping = {
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury"
}
# replace values in education_level
df['vehicle_class'] = df['vehicle_class'].replace(vehicle_mapping)

# Check unique values after mapping
print(df['vehicle_class'].unique())
```
